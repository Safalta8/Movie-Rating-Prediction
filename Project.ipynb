```python
df['Certificate'].unique()
```

```python
convertRatingToLabel = { 'Not Rated': 0,
                        'Unrated': 1,
                        'U': 2,
                        'G': 3,
                        '13': 4,
                        'PG': 5,
                        'M/PG': 6,
                        'PG-13': 7,
                        'UA': 8,
                        'TV-PG': 9,
                        'TV-14': 10,
                        'Approved': 11,
                        'Passed': 12,
                        'GP': 13,
                        'A': 14,
                        'TV-MA': 15,
                        'R': 16,
                        '18': 17,
                        'NC-17': 18 }
```

```python
df['Rating_encoded'] = df['Certificate'].replace(convertRatingToLabel)
df
```

```python
df['Rating_from_10'] = df['Rating_from_10'].fillna(df['Rating_from_10'].median())
```

```python
df.columns
```

```python
X = df[['Rank','Year','Runtime_in_min','Metascore','Gross_in_$_M','Genre_1','Genre_2','Genre_3','Rating_encoded']]
X
```

```python
y = df["Rating_from_10"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred)*100
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model)
```
